```python
#notebook to agroup the datasets

from pathlib import Path
from glob import glob
from scipy.io import loadmat
import pandas as pd

#list with all the data files 
fnames = sorted(glob("data/00-data/*.mat"))
```

```python
#list with all dataframes
df = []
for fname in fnames:
    data = loadmat(fname) #carrega arquivo do tipo matlab 
    var = Path(fname).stem.split('_') #nome do arquivo sem extensão mat
    dates = data['datenum'].squeeze() #valores das datas
    var[-1] = data['values'].squeeze() #valores das variáveis
    df.append (pd.DataFrame([dates, var[-1]], index = ['dates', fname]).T)
```

```python
#merging dataframes to criate only one

ds = pd.merge(df[0], df[1], how = 'left', on = 'dates')
for i in range (2,12):
    ds = pd.merge(ds, df[i-1], how = 'left', on = 'dates') 
ds
```
