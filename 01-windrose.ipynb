# Wind Rose 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from windrose import WindroseAxes

df = pd.read_csv('0-data/simcosta_buoySC-1_2017.csv', index_col = 'datetime')

df_wind = df[['atm_avg_wnd_spd', 'atm_avg_wnd_dir']]

#dataframe wind march to june 2017
df_wind_all = df_wind['2017-03':'2017-07']
```

```python
#WIND ROSE MARCH TO JUNE 2017

plt.rcParams["font.size"] = 16

#Criando o eixo da rosa dos ventos

ax = WindroseAxes.from_ax()

#Plotando os dados

ax.bar(df_wind_all.atm_avg_wnd_dir, df_wind_all.atm_avg_wnd_spd, 
       normed=True, bins=np.arange(0,10,2), 
       opening=0.8, edgecolor='white')

# Inserindo a legenda

## loc se refere  posio da legenda
## Shadow=False para que a caixa da legenda no tenha sombra

lgd = ax.set_legend(title='Wind Speed (knots)', loc=(1.1, 0), shadow=False)
plt.setp(lgd.get_texts(), fontsize=14)

# Definindo quais vo ser os ticks(linhas cinza no grfico)

ax.set_yticks(np.arange(0, 20, step=4))
ax.set_xticklabels(['N', 'NW',  'W', 'SW', 'S', 'SE','E', 'NE'])
ax.set_theta_zero_location('N')
ax.set_theta_direction(-1)

# Inserindo os labels no eixo

#ax.set_yticklabels(np.arange(0, 20, step=4), fontsize='14')

# Inserindo o ttulo do grfico

plt.title("Wind Direction \nSiMCosta Buoy SC-1 \nMarch to June \n2017", y=1.08, fontsize='16')

#Salvando a Rosa dos ventos

plt.savefig('1-output-graphics/simcosta_windrose_marchtojune_2017.png', bbox_extra_artists=(lgd,), bbox_inches='tight')
```
